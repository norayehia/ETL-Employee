```python
import pandas as pd
def get_employees_df():

  return pd.read_csv(
      "https://gist.githubusercontent.com/kevin336/acbb2271e66c10a5b73aacf82"
        "ca82784/raw/e38afe62e088394d61ed30884dd50a6826eee0a8/employees.csv"
  )
def get_departments_df():
  dep_df = pd.read_csv(
      "https://gist.githubusercontent.com/kevin336/5ea0e96813aa88871c20d315b5"
        "bf445c/raw/d8fcf5c2630ba12dd8802a2cdd5480621b6a0ea6/departments.csv"
  )
  dep_df = dep_df.rename(columns={"DEPARTMENT_ID": "DEPARTMENT_IDENTIFIER"})

  return dep_df

employees = get_employees_df()
departments = get_departments_df()
```

```python
#average
mSalary = employees['SALARY'].mean()
#median
mdSalary = employees['SALARY'].median()
print(mSalary)
print(mdSalary)
```

```python
#lower .25and upper quartiles .75 below of an employees' salaries.
lowerquartile = employees['SALARY'].quantile(0.25)
upperquartile = employees['SALARY'].quantile(0.75)
```

```python
print(lowerquartile)
print(upperquartile)
```

#2

```python
# Merge employees with departments on the department identifier
mergeddf = pd.merge(employees, departments, left_on='DEPARTMENT_ID', right_on='DEPARTMENT_IDENTIFIER')
```

```python
mergeddf.head(4)
```

```python
# Group by department name and calculate the average salary
average_salary = mergeddf.groupby('DEPARTMENT_NAME' )['SALARY'].mean().reset_index()
average_salary
```

#3

```python


# Create a column SALARY_CATEGORY ,msalary above is average salary
employees['SALARY_CATEGORY'] = employees['SALARY'].apply(lambda x: 'low' if x < mSalary else 'high')
```

```python
employees.head(3)
```

#4

Please create another column named SALARY_CATEGORY_AMONG_DEPARTMENT with value "low" when the employee salary is lower than average in his / her department and "high" in the other case.

```python
#Group by department name and calculate the average salary
#average_salary = merged_df.groupby('DEPARTMENT_NAME')['SALARY'].mean().reset_index()

# Rename the column in average_salary to avoid conflicts
average_salary = average_salary.rename(columns={'SALARY': 'AVGDEPARTMENTSALARY'})

average_salary
```

```python
# Merge the average salary back into the main DataFrame to merged_df
merged_df = pd.merge(mergeddf, average_salary, on='DEPARTMENT_NAME', how='left')

merged_df
```

```python
# Create the SALARY_CATEGORY_AMONG_DEPARTMENT column
employees['SALARY_CATEGORY_AMONG_DEPARTMENT'] = merged_df.apply(
    lambda row: 'low' if row['SALARY'] < row['AVGDEPARTMENTSALARY'] else 'high', axis=1
)

```
